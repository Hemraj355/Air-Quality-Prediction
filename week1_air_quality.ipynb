```python
# Week 1 - Air Quality Prediction Project

# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load Dataset
# (Download "city_day.csv" from Kaggle and upload it here)
df = pd.read_csv("city_day.csv")

# Step 3: Preview Dataset
print("Dataset Shape:", df.shape)
print("\nFirst 5 Rows:\n", df.head())

# Step 4: Dataset Info
print("\nDataset Info:\n")
df.info()

# Step 5: Basic Statistics
print("\nDescriptive Statistics:\n", df.describe())

# Step 6: Check Missing Values
print("\nMissing Values per Column:\n", df.isnull().sum())

# Step 7: Simple Visualization - AQI Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['AQI'].dropna(), bins=30, kde=True)
plt.title("Distribution of Air Quality Index (AQI)")
plt.xlabel("AQI")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x="AQI_Bucket", data=df, order=df["AQI_Bucket"].value_counts().index, palette="viridis")
plt.title("Distribution of AQI Buckets")
plt.xlabel("Air Quality Category")
plt.ylabel("Count")
plt.show()
```
